## Top 5 Product Categories by Quantity

```python
Top5 = sales.groupby("Product_Category")["Quantity"].sum().sort_values(ascending=False).head(3).reset_index()
sns.barplot(x="Product_Category", y="Quantity", data = Top5)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example Top 5 categories
Top5 = (
    sales.groupby("Product_Category")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# Create barplot
plt.figure(figsize=(8,5))
ax = sns.barplot(x="Product_Category", y="Quantity", data=Top5, palette="Blues_d")
plt.title("Top 5 Product Categories by Quantity")
plt.xticks(rotation=45)

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.0f}',                      # text = bar height
        (p.get_x() + p.get_width()/2., p.get_height()), # position = center top of bar
        ha='center', va='bottom', fontsize=10, color='black'
    )

plt.show()
```

```python
sales["Date"] = pd.to_datetime(sales["Date"])

# Example: Monthly sales trend
monthly_sales = sales.groupby(sales["Date"].dt.to_period("M"))["Total_Amount"].sum().reset_index()
monthly_sales["Date"] = monthly_sales["Date"].dt.to_timestamp()
monthly_sales
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.lineplot(x="Date", y="Total_Amount", data=monthly_sales, marker="o", color="blue")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```
