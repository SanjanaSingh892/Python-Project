```python
import numpy as  np
from scipy import stats
```

```python
scores = [72, 75, 74, 70, 76, 73, 74, 75, 72, 74,
          76, 77, 73, 74, 75, 72, 74, 73, 75, 76,
          74, 72, 73, 75, 74, 76, 73, 74, 75, 72]
```

```python
mu = np.mean(scores)
print(mu)
```

```python
# Perfrom one-sample t-test

t_statistic,  p_value = stats.ttest_1samp(scores,mu)

print("Sample Mean:", np.mean(scores))
print("t-statistics:", t_statistic)
print("p-value:", p_value)


# Decision at alpja = 0.0.5

alpha = 0.05
if p_value > alpha:
    print("Reject H0: The average score is significantly different from 70.")
else:
    print("Fail tp reject H0 : No signficant evidence that the average differ from 70")
    
```

```python
import numpy as np
from scipy import stats

coffee = [2.1, 2.3, 2.5, 2.0, 2.4, 2.6, 2.2, 2.5, 2.3, 2.4,
          2.7, 2.5, 2.3, 2.6, 2.4, 2.5, 2.2, 2.3, 2.4, 2.6,
          2.5, 2.4, 2.3, 2.5, 2.6, 2.4, 2.5, 2.3, 2.4, 2.6,
          2.5, 2.4, 2.3, 2.5, 2.6, 2.4, 2.5, 2.3, 2.4, 2.6,
          2.5, 2.4, 2.3, 2.5, 2.6, 2.4, 2.5, 2.3, 2.4, 2.6]


mu = 2.0

# Perform one-sample t-test
t_statistics, p_value = stats.ttest_1samp(coffee, mu)

print("Sample Mean:", np.mean(coffee))
print("t-statistics:", t_statistics)
print("p-value:", p_value)


# Decision at  alpha = 0.05

alpha = 0.05

if p_value < alpha:
    print("Reject H0 : Average coffe  consumption is significantly  different from 2 cups/day.")

else :
    print("Fail to reject H0: No significant evidence  of differnce.")
    
```

```python
import numpy as np
from scipy import stats

commute_times = [46,47,48,49,50,47,48,46,49,
                 47,48,46,47,49,50,48,47,46,49,
                 47,48,46,47,49,50,48,47,46,49,
                 47,48,47,49,50,48,47,46,49]


mu = 45

t_statistics, p_value = stats.ttest_1samp(commute_times, mu)

print("Sample Mean:", np.mean(commute_times))
print("t-statistics:", t_statistics)
print("p-value:", p_value)


alpha = 0.05
if p_value < alpha:
    print("Reject H0: Commute time is significant different from 45 minutes.")
else:
    print("Fail to reject H0: No significant evidence of differences.")

    
```

```python
# 4: Side-by-Side Comparision 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset

# 1. Histogram with Mean Line
plt.figure(figsize=(8,5))
plt.hist(std['Score'], bins=20, edgecolor='black')
plt.axvline(std['Score'].mean(), color='red', linestyle='dashed', linewidth=2, label=f"Mean = {std['Score'].mean():.2f}")
plt.title("Histogram of Student Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# 2. Boxplot to Identify Outliers
plt.figure(figsize=(6,4))
sns.boxplot(std['Score'])
plt.title("Boxplot of Student Exam Scores")
plt.show()

# 3. Histogram with Density Curve
plt.figure(figsize=(8,5))
sns.histplot(std['Score'], bins=20, kde=True, edgecolor='black')
plt.title("Histogram with Density Curve")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

# 4. Side-by-Side Comparison (First 250 vs Last 250 Students)
std['Group'] = ['First Half']*250 + ['Second Half']*250
plt.figure(figsize=(8,5))
sns.boxplot(x='Group', y='Score', data=std)
plt.title("Comparison of First vs Last 250 Students")
plt.show()

# 5. Violin Plot for Distribution Shape
plt.figure(figsize=(8,5))
sns.violinplot(x='Group', y='Score', data=std, inner='quartile')
plt.title("Violin Plot of Score Distribution by Group")
plt.show()
```
