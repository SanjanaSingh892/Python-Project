```python
1. Hypothesis Testing
Example Question: Do jobs requiring Python pay more than jobs that don’t?

Method: Independent samples t-test
```

```python
from scipy.stats import ttest_ind

python_jobs = read_data[read_data['skills_python']==1]['salary']
non_python_jobs = read_data[read_data['skills_python']==0]['salary']

t_stat, p_val = ttest_ind(python_jobs, non_python_jobs, equal_var=False)
print("t-statistic:", t_stat, "p-value:", p_val)
```

```python
2. Chi-Square Test of Independence
Example Question: Is Python skill requirement independent of industry?

Method: Chi-square test on contingency table
```

```python
import scipy.stats as stats
import pandas as pd

contingency = pd.crosstab(read_data['company_industry'], read_data['skills_python'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print("Chi-square:", chi2, "p-value:", p)
```

```python
3. ANOVA (Analysis of Variance)
Example Question: Do average salaries differ across industries?

Method: One-way ANOVA
```

```python
from scipy.stats import f_oneway

groups = [group['salary'].values for name, group in read_data.groupby('company_industry')]
f_stat, p_val = f_oneway(*groups)
print("F-statistic:", f_stat, "p-value:", p_val)
```

```python
#Null Hypothesis (H₀): There is no difference in average salary between jobs requiring Python and jobs not requiring Python.

Alternative Hypothesis (H₁): Jobs requiring Python have a different average salary than jobs not requiring Python.
```

```python
# sample t-test
```

```python
from scipy.stats import ttest_ind

# Split data into two groups
python_jobs = read_data[read_data['skills_python']==1]['salary']
non_python_jobs = read_data[read_data['skills_python']==0]['salary']

# Perform Welch’s t-test (doesn't assume equal variance)
t_stat, p_val = ttest_ind(python_jobs, non_python_jobs, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_val)
```
