```python
import matplotlib.pyplot as plt

1️⃣ Line Chart — Monthly Sales Trend
```

```python
superstore.columns
```

```python
superstore["Order_Date"] = pd.to_datetime(superstore["Order_Date"], errors="coerce")
```

```python
monthly_sales = superstore.groupby(superstore["Order_Date"].dt.to_period("M"))["Sales"].sum()
```

```python
monthly_sales.plot(kind="line", marker="o", figsize=(10,5), color="teal")
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

```python
# 2️⃣ Bar Chart — Top 10 Products by Profit
```

```python
top_products = superstore.groupby("Product_Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# Option A: Use Sales
top_products = superstore.groupby("Product_Name")["Sales"].sum().sort_values(ascending=False).head(10)

# Option B: Create a mock Profit column (for teaching/demo)
import numpy as np
superstore["Profit"] = superstore["Sales"] * np.random.uniform(0.15, 0.25, size=len(superstore))
top_products = superstore.groupby("Product_Name")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
top_products.plot(kind="bar", figsize=(10,5), color="coral")
plt.title("Top 10 Products by Profit")   # <-- Correct way
plt.ylabel("Profit")
plt.xlabel("Product Name")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Heatmap — Profit Margins Across States
```

```python
subcat_sales = superstore.groupby("Sub_Category")["Sales"].sum().sort_values(ascending=True).tail(10)

subcat_sales.plot(kind="barh", color="orchid")
plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Sub-Category")
plt.show()
```

```python
segment_sales = superstore.groupby("Segment")["Sales"].sum()

segment_sales.plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(6,6))
plt.title("Sales Distribution by Segment")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns

state_profit = superstore.groupby("State")["Profit"].sum().reset_index()

plt.figure(figsize=(12,8))
sns.heatmap(state_profit.pivot_table(values="Profit", index="State"), cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Profit Margins Across States")
plt.show()
```
