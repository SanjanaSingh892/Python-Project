```python
class Students:
    def __init__(self, name, course, percentage):
        self.name = name
        self.course = course
        self.percentage = percentage


    def honors(self):
        return self.percentage > 90


std1 = Students("Alice", "MCA", 88)
std2 = Students("Bob", "MBA", 95)


print(std1.percentage)
std1.honors()
print(std2.percentage)

std1.honors()
std2.honors()
```
